It is based on the Rabobank annual report 2021 foreward (pages 4 to 6), you can find it here: https://media.rabobank.com/m/569cafc747920bd4/original/Rabobank_Annual-Report-2021-pdf.pdf

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# Importing text
text = open('flavio.txt', mode='r', encoding='utf8').read()
```

```python
# Defining stopwords
stopwords = STOPWORDS
len(stopwords)
```

```python
# Adding some letters and words to skip
stopwords.update(['S', 'Eur', 'million', 'billion'])
```

```python
# Defining a mask (optional)
mask = np.array(Image.open('cloud.jpg'))
```

```python
# Creating the WordCloud
wc = WordCloud(
    background_color='orange',
    stopwords=stopwords, # or a list with words to eliminate
    #mask=mask, #to shape the wordcloud
    #contour_width=3,
    normalize_plurals=True, # singular and plural of the same words are counted in one word
    collocations=False, # Whether to include collocations (bigrams) of two words. Ignored if using (from documentation)
    max_words=10000, # maximum nuber of words
    colormap='Blues_r', # to color words
    height=1000,
    width=1000
).generate(text)
```

```python
# Store it
wc.to_file('flavio_wordcloud.png')
```

```python
# to visualize it
plt.figure(figsize=(10,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
